## independendMH&randomMH

$f(x|\theta)~poisson$ 
$f(\theta)~gamma$ 

$f(\theta|x)~gamma$ <- we want to know( it is poropositional to $e^{-13\theta}\theta^{10}$)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
%matplotlib inline

N=10000
def posterior(x):
    return np.exp(-13*x)*np.power(x,10)
```

```python
### first proposal dist is normal distiburition

answer=[]
a1 =0.5

for i in range(N):
    new = np.random.normal(1,0.4) ### proposal dist that's important
    ####print new
    r =  posterior(new)*st.norm.pdf(a1,1.0,0.4)/posterior(a1)*st.norm.pdf(new,1.0,0.4)
    judge = np.random.uniform(0,1)
    ###print judge
    if judge < r and new >0.0:
        ###print judge,r
        answer.append(new)
        a1 =new
```

```python
answer2=[]
a1 = 4.0

for i in range(N):
    a2 = a1 + np.random.normal(0.0,0.2) ### vairance is importnat
    r = posterior(a2)/posterior(a1)
    judge = np.random.uniform(0,1)
    if judge<r and a2 > 0.0:
        answer2.append(a2)
        a1 = a2
```

```python
print len(answer)
print "acceptance rate,%f"%(len(answer)/(N*1.0))
print "expectance %f"%np.mean(answer)
plt.subplot(1,2,1)
plt.plot(answer)
plt.subplot(1,2,2)
plt.hist(answer)
plt.show()
```

```python
print len(answer2)
print "acceptance rate,%f"%(len(answer2)/(N*1.0))
print "expectance %f"%np.mean(answer2)
plt.subplot(1,2,1)
plt.plot(answer2)
plt.subplot(1,2,2)
plt.hist(answer2)
plt.show()
```
