## Gibbs samplers Rao blockwellization

Poisson Count Data

Record the number of passages of individuals per unit time past some sensor.

0: 139,1:128,2:55,3:25,4 or more:13

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

lamb = 3.0
z = np.zeros(10)

z_array=[]
lambda_array=[]
lamb_emp=[]
lamb_emp_rao=[]
for i in range(0,1000):
    z = [np.random.poisson(3) for i in range(0,13)]
    ##print z
    z_array.append(z)
    lamb = np.random.gamma(shape=313+np.sum(z),scale=1/360.0)
    lambda_array.append(lamb)
    ##calcute the empirical_average
    lamb_emp.append(np.mean(lambda_array))
    ## rao-blockwell
    lamb_emp_rao.append(np.mean(313+np.sum(z_array,1))*1/360.0)

plt.title("simle empirical average")
plt.hist(lamb_emp,30,normed=True)
plt.show()
plt.title("rao-blackwell")
plt.hist(lamb_emp_rao,30,normed=True)
plt.show()
plt.plot(lamb_emp,color="red")
plt.plot(lamb_emp_rao)
plt.show()
```
