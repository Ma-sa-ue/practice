```python
import numpy as np
import scipy as sp
import scipy.stats as st
import matplotlib.pyplot as plt 
import collections
%matplotlib inline
```

```python
def get_bata(c,gamma):
    x=[]
    y=[]
    for i in range(40000):
        hoge =c*gamma/(c+i)
        nk = np.random.poisson(hoge)
        hoge2 = np.random.uniform(0,1,nk)
        x.extend(hoge2)
        y.extend(np.random.beta(1,c+i) for j in range(len(hoge2)))
    return np.array([x,y])

plt.figure(figsize=(8,6))
para_list=[[1.0,1.0],[1.0,10.0],[10.0,1.0],[10.0,10.0]]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title("c="+str(para_list[i][0])+",gamma="+str(para_list[i][1]))
    sample=get_bata(para_list[i][0],para_list[i][1])
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.scatter(sample[0],sample[1])
plt.savefig("beta.png")
plt.show()
```

```python
hey = []
for i in range(100):
    sample = get_bata(c=1.0,gamma=5.0)
    hey.append(np.sum(sample[1]))
print np.mean(hey)
plt.hist(hey,bins=10)
plt.show()
```
